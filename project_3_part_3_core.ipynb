```python
import pandas as pd
from sqlalchemy import create_engine
from sqlalchemy.types import *

import warnings
warnings.filterwarnings("ignore")
```

```python
# Create connection string using credentials following this format
# connection = "dialect+driver://username:password@host:port/database"
username = "root"
password = "root" # (or whatever password you chose during mysql installation)
db_name = "tmdb"
connection = f"mysql+pymysql://{username}:{password}@localhost/{db_name}"
```

```python
engine = create_engine(connection)
engine
```

```python
#read in basics file
df_basics = pd.read_csv("Data/title_basics.csv.gz", low_memory = False)
df_basics.head()
```

```python
## create a col with a list of genres
df_basics['genres_split'] = df_basics['genres'].str.split(',')
df_basics
```

```python
#separate the list of genres into new rows
exploded_genres = df_basics.explode('genres_split')
exploded_genres
```

```python
#save a list of all of the unique genres that appear in the column
unique_genres = sorted(exploded_genres['genres_split'].unique())
unique_genres
```

```python
#Create a new title_genres table
title_genres = exploded_genres[['tconst','genres_split']].copy()
title_genres.head()
```

```python
#Create a genre mapper dictionary to replace string genres with integers
genre_ints = range(len(unique_genres))
genre_map = dict(zip(unique_genres, genre_ints))
genre_map
```

```python
#Replace the string genres in title_genres with the new integer ids.
title_genres['genre_id'] = title_genres['genres_split'].map(genre_map)
title_genres = title_genres.drop(columns='genres_split')
title_genres.head()
#df_basics.head()
```

```python
#Convert the genre map dictionary into a dataframe.
genre_lookup = pd.DataFrame({'genre_id': genre_map.values(),                            
                             'genre_name': genre_map.keys()})
genre_lookup.head()
```

```python
#Create a dataype schema for to_sql
#Calculate max string lengths for object columns
key_len = df_basics['tconst'].fillna('').map(len).max()
title_len = df_basics['primaryTitle'].fillna('').map(len).max()
## Create a schema dictonary using Sqlalchemy datatype objects
schema_dict = {
    "tconst": String(key_len+1), 
    "primaryTitle": Text(title_len+1),
    'startYear':Float(),
    'endYear':Float(),
    'runtimeMinutes':Integer()}
schema_dict
```

```python
df_basics.head()
```

```python
#Drop unwanted columns as per instructions
df_basics.drop(columns=['titleType','originalTitle', 'isAdult', 'genres', 'genres_split'], inplace=True)
df_basics.head()
```

```python
# Save to sql with dtype and index=False
df_basics.to_sql('title_basics',engine,dtype=schema_dict,if_exists='replace',index=False)
#Set tconst as primary key
engine.execute('ALTER TABLE title_basics ADD PRIMARY KEY (`tconst`);')
```

```python
## Create a schema dictonary using Sqlalchemy datatype objects
key_len_g = title_genres['tconst'].fillna('').map(len).max()
schema_t_genre = {
    "tconst": String(key_len_g+1), 
    "genre_id": Integer()}
schema_t_genre
```

```python
# Save to sql with dtype and index=False
title_genres.to_sql('title_genres',engine,dtype=schema_t_genre,if_exists='replace',index=False)
```

```python
## Create a schema dictonary using Sqlalchemy datatype objects
key_len_l = genre_lookup['genre_name'].fillna('').map(len).max()
schema_genre = {
    "genre_id": Integer(), 
    "genre_name": String(key_len_l+1)}
schema_genre
```

```python
# Save to sql with dtype and index=False
genre_lookup.to_sql('genres',engine,dtype=schema_genre,if_exists='replace',index=False)
engine.execute('ALTER TABLE genres ADD PRIMARY KEY (`genre_id`);')
```

```python
#read in ratings file
df_ratings = pd.read_csv("Data/title_ratings.csv.gz", low_memory = False)
df_ratings.head()
```

```python
## Calculate max string lengths for object columns
key_len_r = df_ratings['tconst'].fillna('').map(len).max()
## Create a schema dictonary using Sqlalchemy datatype objects
schema_ratings = {
    "tconst": String(key_len_r+1),
    'averageRating':Float(),
    'numVotes':Integer(),}
schema_ratings
```

```python
# Save to sql with dtype and index=False
df_ratings.to_sql('title_ratings',engine,dtype=schema_ratings,if_exists='replace',index=False)
#Set tconst as primary key
engine.execute('ALTER TABLE title_ratings ADD PRIMARY KEY (`tconst`);')
```

```python
#Read in tmdb combined results file (2000 and 2001 data)
df_temp = pd.read_csv("Data/tmdb_results_combined.csv.gz", low_memory = False)
df_temp.head()
```

```python
#save new dataframe with only the needed columns as per instructions
df_tmdb = df_temp[['imdb_id','revenue', 'budget', 'certification']]
df_tmdb.head()
```

```python
#identify redundant/null rows
df_tmdb.loc[df_tmdb['imdb_id'] == '0']
```

```python
df_tmdb.shape
```

```python
#remove null rows from dataframe
df_tmdb = df_tmdb.loc[df_tmdb['imdb_id'] != '0']
df_tmdb.shape
```

```python
df_tmdb.head()
```

```python
## Calculate max string lengths for object columns
key_len_t = df_tmdb['imdb_id'].fillna('').map(len).max()
cert_len = df_tmdb['certification'].fillna('').map(len).max()
## Create a schema dictonary using Sqlalchemy datatype objects
schema_tmdb = {
    "imdb_id": String(key_len_t+1),
    'revenue':Float(),
    'budget':Float(),
    'certification':String(cert_len+1),}
schema_tmdb
```

```python
#Save to sql with dtype and index=False
df_tmdb.to_sql('tmdb_data',engine,dtype=schema_tmdb,if_exists='replace',index=False)
engine.execute('ALTER TABLE tmdb_data ADD PRIMARY KEY (`imdb_id`);')
```
