```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
#import os

import warnings
warnings.filterwarnings("ignore")
```

```python
df_2000 = pd.read_csv("Data/final_tmdb_data_2000.csv.gz", low_memory = False)
df_2000.head()
```

```python
df_2000.shape
```

```python
df_2001 = pd.read_csv("Data/final_tmdb_data_2001.csv.gz", low_memory = False)
df_2001.head()
```

```python
df_2001.shape
```

```python
#concatenate the 2 dataframes into one 
frames = [df_2000, df_2001]
 
df_results = pd.concat(frames, ignore_index=True)
df_results.head()
```

```python
df_results.shape
```

```python
df_results['valid_fin_info'] = (df_results['budget'] > 0) | (df_results['revenue'] > 0)
df_results['valid_fin_info']
```

```python
df_results['valid_fin_info'].value_counts()
```

```python
#How many movies had at least some valid financial information 
ax = sns.countplot(data=df_results, x='valid_fin_info')
ax.set_title(f"Movies with Valid Financial Information", fontweight='bold')
plt.xlabel("Valid Information");
```

```python
#How many movies are there in each of the certification categories (G/PG/PG-13/R)?
df_results['certification'].value_counts()
```

```python
#Visualization
ax = sns.countplot(data=df_results, x='certification')
ax.set_title(f"Movies by Certification Category", fontweight='bold')
plt.xlabel("Certification Category");
```

```python
#What is the average revenue per certification category?
avg_revenue_by_cert = df_results.groupby('certification')['revenue'].mean()
# plot the result
ax=sns.barplot(x = avg_revenue_by_cert.index, y = avg_revenue_by_cert.values)
ax.set_title(f"Average Revenue per Certification Category", fontweight='bold')
plt.xlabel("Certification Category");
```

```python
#What is the average budget per certification category?
avg_budget_by_cert = df_results.groupby('certification')['budget'].mean()

# plot the result
ax=sns.barplot(x = avg_budget_by_cert.index, y = avg_budget_by_cert.values)
ax.set_title(f"Average Budget per Certification Category", fontweight='bold')
plt.xlabel("Certification Category");
```

```python
#save combined dataframe to file
df_results.to_csv("Data/tmdb_results_combined.csv.gz",compression='gzip',index=False)
```
