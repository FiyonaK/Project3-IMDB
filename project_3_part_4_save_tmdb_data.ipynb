```python
import pandas as pd
import os
from sqlalchemy import create_engine
from sqlalchemy.types import *
import glob

import warnings
warnings.filterwarnings("ignore")
```

```python
# Create connection string using credentials following this format
# connection = "dialect+driver://username:password@host:port/database"
username = "root"
password = "root" # (or whatever password you chose during mysql installation)
db_name = "tmdb"
connection = f"mysql+pymysql://{username}:{password}@localhost/{db_name}"
```

```python
engine = create_engine(connection)
engine
```

```python
FOLDER = 'Data/'
file_list = sorted(os.listdir(FOLDER))
file_list
```

```python
q = FOLDER+"*.csv.gz"
print(q)
```

```python
# Use glob.glob to get COMPLETE filepaths
file_list = glob.glob(q)
file_list
```

```python
# Use glob.glob to get COMPLETE filepaths
q = FOLDER+"final_*.csv.gz"
print(q)
file_list = sorted(glob.glob(q))
file_list
```

```python
## use a list comprehension to load in all files into 1 dataframe
df = pd.concat([pd.read_csv(f) for f in file_list])
df
```

```python
# remove ids that are 0
df = df.loc[ df['imdb_id']!='0']
df
```

```python
df = df.reset_index(drop=True)
df
```

```python
## saving the combined csv to disk
df.to_csv(FOLDER+'combined_tmdb_data.csv.gz',compression='gzip',index=False)
```

```python
#Read in tmdb combined results file (2000 and 2001 data)
df_temp = pd.read_csv(FOLDER+'combined_tmdb_data.csv.gz', low_memory = False)
df_temp.head()
```

```python
#save new dataframe with only the needed columns as per instructions
df_tmdb = df_temp[['imdb_id','revenue', 'budget', 'certification']]
df_tmdb.head()
```

```python
#identify redundant/null rows
df_tmdb.loc[df_tmdb['imdb_id'] == '0']
```

```python
df_tmdb.shape
```

```python
#remove rows with no imdb_id from dataframe
df_tmdb = df_tmdb.loc[df_tmdb['imdb_id'] != '0']
df_tmdb.shape
```

```python
df_tmdb.head()
```

```python
#check for rows with null values
df_tmdb.isna().sum()
```

```python
df_tmdb['certification'].value_counts()
```

```python
df_tmdb['certification'] = df_tmdb['certification'].fillna('NR')
df_tmdb['certification'].value_counts()
```

```python
df_tmdb['certification'] = df_tmdb['certification'].replace('Unrated','NR')
df_tmdb['certification'].value_counts()
```

```python
df_tmdb['certification'].isna().sum()
```

```python
df_tmdb.shape
```

```python
#remove rows with 0 revenue, 0 budget and NR certification
rows_to_drop = df_tmdb[(df_tmdb['revenue'] == 0) & (df_tmdb['budget'] == 0) & (df_tmdb['certification'] == 'NR')].index
df_tmdb.drop(rows_to_drop , inplace=True)
df_tmdb.shape
```

```python
## Calculate max string lengths for object columns
key_len_t = df_tmdb['imdb_id'].fillna('').map(len).max()
cert_len = df_tmdb['certification'].fillna('').map(len).max()
## Create a schema dictonary using Sqlalchemy datatype objects
schema_tmdb = {
    "imdb_id": String(key_len_t+1),
    'revenue':Float(),
    'budget':Float(),
    'certification':String(cert_len+1),}
schema_tmdb
```

```python
#Save to sql with dtype and index=False
df_tmdb.to_sql('tmdb_data',engine,dtype=schema_tmdb,if_exists='replace',index=False)
engine.execute('ALTER TABLE tmdb_data ADD PRIMARY KEY (`imdb_id`);')
```
